Untuk bluetooth yang sering lepas ketika sudah connect ke earphone 

1. Buka config bluetooth : <br>
sudo nano /etc/bluetooth/main.conf

2. Cari bagian [General] <br>
Experimental = true <br>


3. Cari bagian [Policy] <br>
AutoEnable = true <br>
FastConnectable = true

4. Matikan Power Mangement Bluetooth <br>
echo "options btusb enable_autosuspend=n" | sudo tee /etc/modprobe.d/bluetooth-no-autosuspend.conf

Jika sudah restart bluetooth <br>
sudo systemctl restart bluetooth
