untuk buat gesture swipe dari kiri ke kanan 2 jari agar bisa backward kita buatkan confignya untuk browser tersebut (basis chromium)

1. Buat file config (chrome ini bisa diganti entah itu brave atau browser basis chrome lainnya) <br>
nano ~/.config/chrome-flags.conf

2. tambahkan config <br>
--enable-features=TouchpadOverscrollHistoryNavigation<br>
--ozone-platform-hint=auto   # Auto ini bisa diganti ke wayland
