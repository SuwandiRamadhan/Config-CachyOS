# Optimasi Jaringan Linux untuk Gaming (CachyOS/Arch)

### Masalah utama
Pengaturan default jaringan di Linux seringkali dioptimalkan untuk **hemat daya** (*power saving*) dan **kestabilan standar** (algoritma Cubic), yang menyebabkan:
* **Jitter Tinggi :** Kartu Wi-Fi masuk mode "tidur" secara berkala saat trafik rendah.
* **Lag Spikes :** Algoritma "Cubic" memotong kecepatan secara drastis saat mendeteksi *packet loss* kecil.

### Langkah Konfigurasi
1. Matikan Wi-Fi Power Saving <br>
Langkah ini mencegah kartu Wi-Fi masuk ke mode tidur (*sleep*) saat ada jeda trafik mikro, sehingga menghilangkan lonjakan ping yang tiba-tiba.
<br><br>
**File :** `/etc/NetworkManager/conf.d/default-wifi-powersave-on.conf`.
<br>
Isi dengan : <br>
[connection]  <br>
// 2 = Matikan Power Saving (Performa Maksimal) <br>
// 3 = Aktifkan Power Saving (Default, sering menyebabkan lag) <br>
wifi.powersave = 2

<br><br>
* **Kalau sudah save :** <br>
CTRL + S <br>
CTRL + Q

<br><br>
* **Terapkan Perubahan :** <br>
sudo systemctl restart NetworkManager

<br><br>
* **Verifikasi :** <br>
// Output harus menampilkan status "Power Management:off"
iwconfig
<br><br>

2. Aktifkan BBR & CAKE
Mengubah kontrol kemacetan (congestion control) TCP dari Cubic ke BBRv3 (algoritma Google) dan menggunakan CAKE untuk manajemen antrian paket.

* BBR: Mengabaikan packet loss minor dan tetap menjaga throughput tinggi (ping jauh lebih stabil di Wi-Fi).

* CAKE: Manajemen antrian modern yang memperhalus aliran trafik data agar paket game tidak terjebak antrian.

**File :** /etc/sysctl.d/99-bbr.conf <br>
// Gunakan CAKE untuk queue discipline (lebih baik dari fq_codel untuk gaming) <br>
net.core.default_qdisc = cake <br>
<br>
// Gunakan BBR untuk congestion control <br>
net.ipv4.tcp_congestion_control = bbr <<<< (pakai yang bbr) <br>

<br><br>
* **Terapkan Perubahan :** <br>
// Reload pengaturan sysctl <br>
sudo sysctl --system

<br><br>
* **Verifikasi :** <br>
// Output harus menampilkan "bbr" <br>
sysctl net.ipv4.tcp_congestion_control
<br><br>
